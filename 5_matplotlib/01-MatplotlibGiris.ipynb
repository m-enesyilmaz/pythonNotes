```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# bu komut diğer editörerde çalışmaz jupyter için yazdık
```

```python
yasListesi = [12,20,30,30,40,50,60,70,75,85]
kiloListesi = [20,60,80,85,68,97,80,76,75,100]
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"g")
plt.xlabel("Yas")
plt.ylabel("Kilo")
# plt.show() yok çünkü %matpoltlib inline komutu çalıştırdık
plt.title("Kilo'nun yaşa Göre Değişim Grafiği")
```

```python
numpyDizisi = np.linspace(0,10,20)
numpyDizisi
```

```python
numpyDizisi2 = numpyDizisi ** 3
numpyDizisi2
```

```python
plt.plot(numpyDizisi,numpyDizisi2,"r*-") # -- gibi görselleştirmeler yapabiliriz
```

```python
plt.subplot(1,2,1) # satır, sütun, hangisi
plt.plot(numpyDizisi,numpyDizisi2,"b*-")
plt.subplot(1,2,2) # 1 sıra olacak, 2 kolon olacak, şu an 2. grafiği çiziyorum
plt.plot(numpyDizisi2,numpyDizisi,"g--")
```

```python
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.2, 0.2, 0.9, 0.9])
# figure x başlama yeri,figure y başlama yeri, width, hight
figureAxes.plot(numpyDizisi,numpyDizisi2,"g")
figureAxes.set_xlabel("X Ekseni Veri İsmi")
figureAxes.set_ylabel("Y Ekseni Veri İsmi")
figureAxes.set_title("Grafik Başlığı")
```

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1, 0.1, 0.9, 0.9])
eksen2 = figure2.add_axes([0.2, 0.6, 0.3, 0.3])

eksen1.plot(numpyDizisi,numpyDizisi2,"g")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("Ana Grafik Başlık")

eksen2.plot(numpyDizisi2,numpyDizisi,"b")
eksen2.set_xlabel("X Ekseni Küçük")
eksen2.set_ylabel("Y Ekseni Küçük")
eksen2.set_title("Küçük Grafik Başlık")

# eksen1 = figure2.add_axes([0.1, 0.1, 0.7, 0.7])
# eksen2 = figure2.add_axes([0.3, 0.3, 0.3, 0.3])
# bu değerlerle çalıştırdığımızda göreceğiz ki üst üste biniyorlar bunları değiştirerek ayarlasını yapıyoruz
```

```python
(benimFigure_M, benimEksen_M) = plt.subplots(nrows=1, ncols=2) # subplots tuple döndürüyor aslında
type(benimEksen_M)

for eksen in benimEksen_M:
    eksen.plot(numpyDizisi,numpyDizisi2,"g")
    eksen.set_xlabel("X Ekseni")
    eksen.set_ylabel("Y Ekseni")
    eksen.set_title("Title")
    
plt.tight_layout()
```

## görsel geliştirmeler

```python
yeniFigure = plt.figure()
# figure size ıı buradan da ayarlanabiliyor.
# örneğin; yeniFigure = plt.figure(figsize=(6,6))
# dpi ile figure mizin kalitesini arttırıyoruz
# yeniFigure = plt.figure(dpi = 120) gibi
yeniEksen = yeniFigure.add_axes([0.1, 0.1, 0.9, 0.9])
yeniEksen.plot(numpyDizisi,numpyDizisi ** 2, label = "numpyDizisi ** 2")
yeniEksen.plot(numpyDizisi, numpyDizisi ** 3 ,label = "numpyDizisi * 3")
yeniEksen.legend(loc = 6) # loc ile legend locasyonu seçebiliyoruz
```

```python
yeniFigure.savefig("benimFigure.png",dpi= 200)
```
