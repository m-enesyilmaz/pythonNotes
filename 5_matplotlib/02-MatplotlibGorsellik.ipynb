```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
numpyDizisi1 = np.linspace(0,10,20)
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 2
```

```python
(benimFigur, benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1, numpyDizisi2, color = "#6495ED")
benimEksen.plot(numpyDizisi2, numpyDizisi1, color = "#E6E6FA")
# eğer transparanlık katmak istersek alpha değerini değiştirmek gerekiyor
# örneğin; benimEksen.plot(numpyDizisi1, numpyDizisi2, color = "#6495ED", alpha = 0.3) gibi
benimEksen.plot(numpyDizisi2, np.sqrt(numpyDizisi1 ** 3), color = "#9932CC", alpha = 0.3)
```

```python
(yeniFigure2, yeniEksen2) = plt.subplots(dpi = 120)
yeniEksen2.plot(numpyDizisi1,numpyDizisi1 + 2, color = "#000080", linewidth = 1.5)
yeniEksen2.plot(numpyDizisi1, numpyDizisi1 + 3, color = "yellow",  linewidth = 0.5)

yeniEksen2.plot(numpyDizisi1, numpyDizisi1 + 4, color = "#FFDAB9", linestyle = "-.")
yeniEksen2.plot(numpyDizisi1, numpyDizisi1 + 5, color = "#008000", linestyle = ":")
yeniEksen2.plot(numpyDizisi1, numpyDizisi1 + 6, color = "#B22222", linestyle = "--")

yeniEksen2.plot(numpyDizisi1, numpyDizisi1 + 10, color = "#696969", linestyle = "--", marker = "o", markersize = "4", markerfacecolor = "red")
yeniEksen2.plot(numpyDizisi1, numpyDizisi1 + 12, color = "#696969", linestyle = "--", marker = "+", markersize = "8")
```
