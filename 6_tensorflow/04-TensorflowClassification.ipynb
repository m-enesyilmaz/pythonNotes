```python
# x ve y mizi tekrar belirliyoruz
y = dataFrame["Type"].values
x = dataFrame.drop("Type", axis= 1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.3, random_state= 15)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
x_train.shape
```

```python
model = Sequential()

# 30 nöron başlamamızın sebebi train verimizde 30 feature olamasında dolayı x_train --> (383,30)
# her zaman böylemi midir hayır ama bunu deneyerek başlamak neden olmasın 
model.add(Dense(units= 30, activation= "relu"))
model.add(Dense(units= 15, activation= "relu"))
model.add(Dense(units= 15, activation= "relu"))
model.add(Dense(units= 1, activation= "sigmoid"))
# burada 30 başlayıp ekleyeceğimiz katman boyunca azaltmakta ilk denemeler için tercih ediliyor
# sonra değiştiriliyor fakat genel geçer bir deneme diyebiliriz
# biz burada sınıflandırma yapacağımız için son çıkış katmanının aktivasyon fonksiyonunu 
# sigmoid olarak belirliyoruz.

model.compile(loss="binary_crossentropy", optimizer= "adam")
# normalde loss fonkisyonumuzda mse kullanıyorduk ama sınıflandırma yaptığımız için loss fonksiyonumuzu
# binary_crossentropy olarak seçtik.
```

```python
model.fit(x= x_train, y= y_train, epochs= 700, validation_data=(x_test, y_test), verbose = 1)
# burada epoch değerimizi çok abartılı vermek bizim modelimizin overfitting olmasına sebep olabiliyor
# yeni modelimiz öğrenmek yerine ezberlemeye başlıyor
```

```python
model.history.history
```

```python
modelKaybi = pd.DataFrame(model.history.history)
modelKaybi.plot()
```

```python
# burada da görüleceği üzere validation loss overfitting oluyor ve alıp başını gidiyor xd
# yahşi günde yar yahşidir yaman günde yetiş gardaş xd
```

```python
# epoch u bir yerde durdurabiliyoruz bunun önüne geçebilmek için
```

```python
# modelimizi üstten getiriyoruz; 
model = Sequential()

model.add(Dense(units= 30, activation= "relu"))
model.add(Dense(units= 15, activation= "relu"))
model.add(Dense(units= 15, activation= "relu"))
model.add(Dense(units= 1, activation= "sigmoid"))

model.compile(loss="binary_crossentropy", optimizer= "adam")
```

```python
earlyStopping = EarlyStopping(monitor= "val_loss", mode= "min", verbose= 1, patience = 25)
# burada val_loss u takip et dedik, modumuz minimumda olacak, verbose biraz izle, 
# eopch da bir değişim olmaması durumunun takibi için ise patience
```

```python
model.fit(x= x_train, y= y_train, epochs= 700, validation_data=(x_test, y_test), verbose= 1, callbacks= [earlyStopping])
# burada diğer eğitimden farklı olarak callbacks çağırdık bunu biz dizi halinde istiyor bizden 1 tane olmasına rağmen
```

```python
modelKaybi2 = pd.DataFrame(model.history.history)
modelKaybi2.plot()
```

```python
# bunu yapmamıza rağmen hala burada memnun değilsek yapmamız gereken dropout kavramı
```

```python
# modelimizi üstten getiriyoruz dropout örneği için; 
model = Sequential()

model.add(Dense(units= 30, activation= "relu"))
model.add(Dropout(0.5))
# burada istediğimiz ilk katmanın sonuna gelip yüzde kaçında rastgele bir şekilde
# dropout yapacağımıza karar vermemiz gerekiyor. genellikle 0.5 ile denenir ama 
# 0.5 üstü denendiğinde sistem uyarı verir. Genellikle denenilen 0.1 ile 0.5 arasında 
# bir durumdur. Her katmandan sonra ekliyoruz

model.add(Dense(units= 15, activation= "relu"))
model.add(Dropout(0.5))

model.add(Dense(units= 15, activation= "relu"))
model.add(Dropout(0.5))

model.add(Dense(units= 1, activation= "sigmoid"))

model.compile(loss="binary_crossentropy", optimizer= "adam")
```

```python
model.fit(x= x_train, y= y_train, epochs= 700, validation_data=(x_test, y_test), verbose= 1, callbacks= [earlyStopping])
```

```python
kayipDF = pd.DataFrame(model.history.history)
```

```python
kayipDF.plot()
```

```python
# overfitting in biraz üstesinden geldik gibi görünüyor bir de tahminlerimize bakarsak;
# tahminlerimiz = np.argmax(model.predict(x_test), axis= -1)
# classes_x = np.argmax(tahminlerimiz,axis = 1)
# classes_x
# tahminlerimiz
predict_x = model.predict(x_test) 
classes_x = np.argmax(predict_x,axis=1)
classes_x
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, classes_x))
```

```python
print(confusion_matrix(y_test, classes_x))
```
