```python
dataFrame = dataFrame.drop("transmission", axis= 1)
# transmission verimizde kullanmayacağız o yüzden düşürüyoruz
```

```python
# şimdi ise ulaşmak istediğimizi bulacağız yani x ve y mizi
y = dataFrame["price"].values
x = dataFrame.drop("price", axis= 1).values # x ise price hariç hepsi dedik
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.3, random_state= 10)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# burada x_train içerisinde 5 feature var bu yüzden min 5 katmandan başlatmak daha mantıklı olabilir
x_train.shape
```

```python
model = Sequential()

#               12 nörün ve activasyon fonk ReLU
model.add(Dense(12, activation= "relu"))
model.add(Dense(12, activation= "relu"))
model.add(Dense(12, activation= "relu"))
model.add(Dense(12, activation= "relu"))

model.add(Dense(1)) # çıkış katmanı ekledik 1 tane

model.compile(optimizer= "adam", loss= "mse")
```

```python
model.fit(x = x_train, y = y_train, validation_data=(x_test, y_test), batch_size = 250, epochs= 300)
# burada doğrulama datamızı tuple olarak verdik ve batch size belirledik diğer örnekten farklı olarak
```

```python
kayipVerisi = pd.DataFrame(model.history.history)
kayipVerisi.head()
```

```python
# plt.figure(figsize=(7,5))
kayipVerisi.plot()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
tahminDizisi = model.predict(x_test)
tahminDizisi
```

```python
mean_absolute_error(y_test, tahminDizisi)
```
